```python
import gnwrapper
import gym

env = gnwrapper.Animation(gym.make('LunarLander-v2', render_mode="rgb_array"))

obs = env.reset()

for _ in range(1000):
    next_obs, reward, term, trunc, info = env.step(env.action_space.sample())
    env.render()

    obs = next_obs
    if term or trunc:
        obs = env.reset()
```

```python
import gnwrapper
import gym

env = gnwrapper.LoopAnimation(gym.make('LunarLander-v2', render_mode="rgb_array"))

obs = env.reset()

for _ in range(200):
    next_obs, reward, term, trunc, info = env.step(env.action_space.sample())
    env.render()

    obs = next_obs
    if term or trunc:
        obs = env.reset()

env.display()
```
